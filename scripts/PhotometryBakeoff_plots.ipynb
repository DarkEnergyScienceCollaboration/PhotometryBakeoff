```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.ion()
plt.rcParams['figure.figsize'] = (15, 10)

# Read in the baker results as numpy records.
names = 'ug gr ri iz zy sedname redshift time'.split()
lsst = np.recfromtxt('../results/lsst_bakeoff_output.txt', names=names)
photozdc1 = np.recfromtxt('../results/photozdc1_bakeoff_output.txt', names=names)
galsim = np.recfromtxt('../results/galsim_bakeoff_output.txt', names=names)
```

```python
# Check that the sednames match (unmangling for the PhotoZDC1 names).
for i in range(len(lsst['sedname'])):
    if lsst['sedname'][i] != photozdc1['sedname'][i].replace('_', '.'):
        print "photozdc1 name mismatch:", lsst['sedname'][i], photozdc1['sedname'][i]
    if lsst['sedname'][i] != galsim['sedname'][i]:
        print "galsim name mismatch:", lsst['sedname'][i], galsim['sedname'][i]
```

```python
# Plot photozdc1 vs lsst and galsim vs lsst.
fig = plt.figure()
subplots = [231, 232, 233, 234, 235]
for color, subplot in zip(names[:5], subplots):
    axes = fig.add_subplot(subplot)
    handles = []
    handles.append(plt.scatter(lsst[color], photozdc1[color]-lsst[color], label='PhotoZDC1'))
    handles.append(plt.scatter(lsst[color], galsim[color]-lsst[color], alpha=0.3, color='green', marker='o',
                               label='GalSim'))
    plt.legend(handles=handles, scatterpoints=1, loc=2)
    plt.xlabel('sims_photUtils')
    plt.ylabel('[PhotoZDC1,GalSim] - sims_photUtils')
    axes.set_title('-'.join(color))
```

```python
# Plot histograms of photozdc1[color]-lsst[color] and galsim[color]-lsst[color].
def hist_range(x, y):
    xy = np.concatenate((x, y))
    return min(xy), max(xy)

fig = plt.figure()
for color, subplot in zip(names[:5], subplots):
    axes = fig.add_subplot(subplot)
    dphotozdc1 = photozdc1[color] - lsst[color]
    dgalsim = galsim[color] - lsst[color]
    x_range = hist_range(dphotozdc1, dgalsim)
    plt.hist(dphotozdc1, label='PhotoZDC1 - sims_photUtils', histtype='step', bins=30, range=x_range)
    plt.hist(dgalsim, alpha=0.5, color='green', label='GalSim - sims_PhotUtils', 
             histtype='step', bins=30, range=x_range)
    plt.legend()
    plt.xlabel('[PhotoZDC1,GalSim] - sims_photUtils')
    axes.set_title('-'.join(color))
```

```python
# Histogram the execution times.
fig = plt.figure()
plt.hist(np.log10(lsst['time']), histtype='step', color='black', range=(-2.5, -1), bins=50, label='sim_photUtils')
plt.hist(np.log10(photozdc1['time']), histtype='step', color='blue', range=(-2.5, -1), bins=50, label='PhotoZDC1')
plt.hist(np.log10(galsim['time']), histtype='step', color='green', range=(-2.5, -1), bins=50, label='GalSim')
print 'log10(lsst) range:', min(np.log10(lsst['time'])), max(np.log10(lsst['time']))
print 'log10(photozdc1) range:', min(np.log10(photozdc1['time'])), max(np.log10(photozdc1['time']))
print 'log10(galsim) range:', min(np.log10(galsim['time'])), max(np.log10(galsim['time']))
plt.legend()
plt.xlabel('log10(execution time / s)')
```
